# Filtering with Mean & Gaussian Filters


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#using car.jpg

image = cv2.imread('car.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
figure_size = 9

#Mean Filter
mean_image = cv2.blur(image,(figure_size, figure_size))
plt.figure()
plt.imshow(cv2.cvtColor(mean_image, cv2.COLOR_HSV2RGB)), plt.title('Mean Filter')
plt.savefig('res01.jpg')

#Gaussian Filter
gauss_image = cv2.GaussianBlur(image, (figure_size, figure_size), 5)
plt.figure()
plt.imshow(cv2.cvtColor(gauss_image, cv2.COLOR_HSV2RGB)), plt.title('Gaussian Filter')
plt.savefig('res02.jpg')

plt.show()
```
