# Denoising with Median Filter
For figure_size $\le$ 5 images will be too noisy and also for figure_size $\ge$ 9 images will be too dim.
Accoeding to reasons we had talk about I use figure_size = 7 for filtering.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#read images
image = cv2.imread('noise.jpg')
image1 = cv2.imread('noise1.jpg')

#size
figure_size = 7

#convet to HSV
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2HSV)

#apply filters
image = cv2.medianBlur(image, figure_size)
image1 = cv2.medianBlur(image1, figure_size)

#ploting outputs
plt.figure()
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB))
plt.subplot(122), plt.imshow(cv2.cvtColor(image1, cv2.COLOR_HSV2RGB))

#saving figure
plt.savefig('res03.jpg')

plt.show()
```
